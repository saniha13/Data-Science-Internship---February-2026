```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ").upper()

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()

light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

employee = Employee(name, emp_id, dept)
employee.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Input
name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))


# Input
product_name = input("Product Name: ")
price = input("Price: ")

product = Product(product_name, price)
product.display_price_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Input
movie_name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(movie_name, rating)
movie.display_movie()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):
        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Input
customer_name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(customer_name, address)
delivery.display_delivery()
```
