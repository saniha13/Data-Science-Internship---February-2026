```python
def validate_recharge(amount, valid_plans):
    # Check minimum recharge amount
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    # Check if plan is valid
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False

    return True


# Valid recharge plans
valid_plans = [199, 299, 399, 599]

# Retry loop
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(amount, valid_plans):
            print("Recharge successful!")
            break   # Exit loop if valid
        else:
            print("Please try again.\n")

    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

```python
def check_inventory(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"Reorder Alert: {product} stock is low ({quantity} left)")
        else:
            print(f"Stock OK: {product} has sufficient stock ({quantity})")


# Dictionary of products and their stock quantities
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 5,
    "Soap": 18
}

# Function call
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    # Determine result
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Input list of student marks
marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    mark = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)

# Function call
average, result = process_result(marks)

print(f"\nAverage Marks: {average:.2f}")
print("Result:", result)
```

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra for peak hour
    if is_peak:
        fare += fare * 0.25

    return fare


# Retry loop
while True:
    try:
        distance = float(input("Enter distance travelled (in km): "))

        peak_input = input("Is it peak hour? (yes/no): ").lower()
        is_peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, is_peak)

        print(f"\nTotal Cab Fare: ₹{total_fare:.2f}")

        retry = input("\nDo you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break

    except ValueError:
        print("Please enter valid numeric distance.\n")
```

```python
def check_eligibility(attendance_list):
    present_count = 0

    # Count present days using loop
    for day in attendance_list:
        if day.upper() == "P":
            present_count += 1

    total_days = len(attendance_list)
    attendance_percentage = (present_count / total_days) * 100

    if attendance_percentage >= 75:
        return attendance_percentage, "Eligible"
    else:
        return attendance_percentage, "Not Eligible"


# Input attendance
attendance = []

n = int(input("Enter total number of working days: "))

for i in range(n):
    status = input(f"Enter attendance for day {i+1} (P/A): ")
    attendance.append(status)

# Function call
percentage, result = check_eligibility(attendance)

print(f"\nAttendance Percentage: {percentage:.2f}%")
print("Eligibility Status:", result)
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_characters = "@#$"

    # Check each character using loop
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True

    # Check conditions
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Input from user
password = input("Enter your password: ")

# Function call
result = check_password_strength(password)

print(result)
```
