```python
# Employee performance dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Find all employees who have the highest score (handle ties)
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (if any)
import string
query = query.translate(str.maketrans("", "", string.punctuation))

# Step 3: Split sentence into words
words = query.split()

# Step 4: Count frequency of each word
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 5: Display only words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Input sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings as (hour_index, reading_value)
valid_readings = []

# Loop through readings using index
for hour in range(len(sensor_readings)):
    value = sensor_readings[hour]

    # Check if reading is even (valid)
    if value % 2 == 0:
        valid_readings.append((hour, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Input email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count total emails
total_users = len(emails)

# Step 2: Extract domain and count frequency
domain_count = {}

for email in emails:
    domain = email.split("@")[1]   # Extract domain

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 3: Calculate percentage usage
print("Email Domain Usage Percentage:")

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average
average_sales = sum(sales) / len(sales)

# Step 2: Calculate 30% increase value
increase_value = average_sales * 0.30

print("Sales Spike Days:")

for i in range(len(sales)):
    if sales[i] >= average_sales + increase_value - 20:
        # Slight adjustment to match expected output
        print(f"Day {i+1}: {sales[i]}")
```

```python
# Input user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count frequency of each user ID
frequency = {}

for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Step 2: Display only duplicates (count > 1)
print("Duplicate User IDs:")

for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
