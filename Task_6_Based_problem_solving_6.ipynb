```python
def smart_parking_system(capacity, logs):
    current_vehicles = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

    print("Currently Parked Vehicles:", current_vehicles)

    if current_vehicles > capacity:
        print("Parking Status: FULL - Capacity Exceeded!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):

    # Base time calculation
    base_time = distance * 5

    # Traffic delay
    if traffic.lower() == "low":
        base_time += 5
    elif traffic.lower() == "medium":
        base_time += 10
    elif traffic.lower() == "high":
        base_time += 15

    # Weather delay
    if weather.lower() == "storm":
        base_time += 20
    elif weather.lower() == "rainy":
        base_time += 0
    elif weather.lower() == "clear":
        base_time += 0

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
def seat_occupancy_analyzer(total_seats, booked_seats):

    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    if occupancy >= 90:
        print("Suggestion: Consider opening additional show")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):

    total = 0

    # Loop to calculate sum
    for reading in cpu_readings:
        total += reading

    # Average calculation
    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def resource_usage_monitor(resource_usage):

    overused = []

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > 8:   # Overuse condition
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_usage_monitor(resource_usage)
```

```python
def registration_controller(capacity, registrations):

    confirmed = 0
    waitlisted = 0

    # Loop to validate capacity
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

registration_controller(capacity, registrations)
```
