```python
# Stored credentials
username = "admin"
password = "1234"

# User input
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Login check
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Number of Students Passed:", pass_count)
print("Number of Students Failed:", fail_count)
```

```python
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# Cleaned list
cleaned_names = []

for name in names:
    # Remove leading/trailing spaces and convert to lowercase
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print the cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Status: Long message (more than 10 characters)")
    else:
        print("Status: Short message")

    print("-" * 30)

```

```python
# List of system log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize error counter
error_count = 0

# Count ERROR entries
for entry in logs:
    if entry == "ERROR":
        error_count += 1

# Print total error count
print("Total ERROR entries:", error_count)
```
