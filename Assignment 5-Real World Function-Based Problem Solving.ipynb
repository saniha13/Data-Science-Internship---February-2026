```python
def analyze_engagement(likes_list):
    total_likes = 0

    # Using loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    # Conditional logic to check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Display results
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example data (likes received at different times)
likes_data = [250, 300, 150, 200, 150]

analyze_engagement(likes_data)
```

```python
def check_medicine_stock(stock):

    # Conditional logic to check stock level
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    # Display results
    print("Medicine Stock:", stock)
    print("Status:", status)


# Example usage
current_stock = 6
check_medicine_stock(current_stock)
```

```python
def check_rainfall(rainfall_data, required_level):

    total_rainfall = 0

    # Using loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain

    # Mathematical operation to calculate average
    average_rainfall = total_rainfall / len(rainfall_data)

    # Conditional logic to check adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Display results
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Example seasonal rainfall data (in mm)
season_data = [70, 75, 80, 65, 70]
required_rainfall = 70

check_rainfall(season_data, required_rainfall)
```

```python
def detect_duplicate_accounts(usernames):

    unique_users = set()
    duplicates_found = False

    # Loop through usernames
    for user in usernames:
        if user in unique_users:
            duplicates_found = True
            break
        else:
            unique_users.add(user)

    # Conditional output
    if duplicates_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example username list
username_list = ["saniha01", "tech_girl", "coder123", "saniha01"]

detect_duplicate_accounts(username_list)
```

```python
def filter_premium_crops(price_list):

    premium_crops = []

    # Loop and list filtering logic
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    # Display result
    print("Premium Crops:", premium_crops)


# Example crop prices (in ₹)
crop_prices = [1500, 2500, 1800, 3200, 1999]

filter_premium_crops(crop_prices)
```

```python
def check_application_health(error_count):

    # Conditional statements to evaluate system status
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Display results
    print("Error Count:", error_count)
    print("System Status:", status)


# Example usage
errors = 7
check_application_health(errors)
```

```python
def check_transaction_limit(amount):

    daily_limit = 50000  # ₹50,000 limit

    # Conditional logic to validate transaction
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    # Display results
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example usage
transaction_amount = 60000
check_transaction_limit(transaction_amount)
```

```python
def check_exam_eligibility(attendance_list):

    total_classes = len(attendance_list)
    attended_classes = 0

    # Using loop to count attended days
    for status in attendance_list:
        if status == "P":   # P = Present
            attended_classes += 1

    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    # Conditional check for eligibility
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    # Display results
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", eligibility)


# Example attendance data
attendance_data = ["P", "P", "A", "P", "P"]  # 4 out of 5 present = 80%
check_exam_eligibility(attendance_data)
```

```python
def calculate_electricity_bill(units):

    bill = 0

    # Slab-based billing calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


# Example usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Classification:", status)
```
